I decided to do the same things as from the training set, but instead of deleting the lines, I replaced them with vector averages

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('test.csv')
df.Date = pd.to_datetime(df.Date)
df = df.sort_values(by = 'Date')
df
```

```python
df.describe()
```

```python
sns.boxplot(df.Ask_Price)
```

```python
indexes = df[df.Ask_Price > 50000].index

for i in indexes:
    df.at[i, 'Ask_Price'] = np.mean(df.Ask_Price)
```

```python
df.describe()
```

```python
sns.boxplot(df.FeaturePrice1)
```

```python
indexes = df[df.FeaturePrice1 > 50000].index

for i in indexes:
    df.at[i, 'FeaturePrice1'] = np.mean(df.FeaturePrice1)
```

```python
df.describe()
```

```python
sns.boxplot(df.FeaturePrice3)
```

```python
indexes = df[df.FeaturePrice1 > 50000].index

for i in indexes:
    df.at[i, 'FeaturePrice3'] = np.mean(df.FeaturePrice1)
```
